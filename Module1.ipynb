```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
data.dtypes
```

```python
data['Date'].describe
```

```python
init_date=data['Date'].min()
```

```python
last_date=data['Date'].max()
```

```python
duration=last_date-init_date
print(duration)
```

```python
#printing last 90 days
n_days= last_date-timedelta(90)
print(n_days)
```

```python
#summary of close price of last 90 days|
data[data['Date']>=  n_days]['Close Price'].describe()
```

```python
#summary of last 90 transaction
three_month_summary=data['Close Price'].tail(90).describe()
print(three_month_summary) # 
```
